```python
def load_and_preprocess_data(file_path, text_variables, num_variables, target_variable):

    df = pd.read_csv(file_path)

    df.columns = df.columns.str.strip().str.replace('  ', ' ')

    print("Cleaned Column Names:\n", df.columns)
    
    X = df[text_variables + num_variables]  
    y = df[target_variable]
    
    return X, y
```
